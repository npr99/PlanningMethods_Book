# 1. UNDERSTANDING THE ACS: THE BASICS
## What is the ACS?
The American Community Survey (ACS) is a nationwide survey designed to provide communities with reliable and timely social, economic, housing, and
demographic data every year

* The ACS has an annual sample size of about 3.5 million addresses
* Estimates reflect data that have been collected over a period of time
* ACS 1-year estimates are available for geographic areas with at least 65,000 people.
* ACS 5-year estimates are available for all geographic areas 

Because the ACS is based on a sample, rather than all housing units and people, ACS estimates have a degree of uncertainty associated with them, called sampling error. In general, the larger the sample, the smaller the
level of sampling error. To help users understand the impact of sampling error on data reliability, the Census Bureau provides a “margin of error” for each published ACS estimate. The margin of error, combined with the
ACS estimate, give users a range of values within which the actual “real-world” value is likely to fall.

## How are ACS Data Collected?
The ACS collected data using 
1. Internet response
2. Paper questionnaires through the mail
3. Personal visits with a Census Bureau interviewer

The annual sample size of the ACS has increased over
time, from 2.9 million addresses in 2005 to more than
3.5 million addresses in 2015. 

* The survey is not mailed to specific people, but rather to specific addresses
*  The Census Bureau selects a random sample of addresses to be included in the ACS. 
* Each address has about a 1-in-480 chance of being selected in a given month
* No address should be selected more than once every 5 years
* Over a 5-year period, the Census Bureau samples approximately 1-in-9 households nationwide
* The sampling rate is higher in areas with small populations 
* The sampling rate is higher in areas with low predicted response rates

# Replication code for Figure 3.2. Child Poverty Rate in Rice County, Minnesota: 2015 and 2011-2015.

## Step 1: Obtain Data

```
# Python packages required to read in and Census API data
import requests ## Required for the Census API
import pandas as pd # For reading, writing and wrangling data
```

```
apijson = requests.get('https://api.census.gov/data/2015/acs/acs1/profile?get=GEO_ID,LSAD_NAME,DP03_0129PE,DP03_0129PM&in=state:27&for=county:131')
# Convert the requested json into pandas dataframe
acs1year = pd.DataFrame(columns=apijson.json()[0], data=apijson.json()[1:])
acs1year.head()
```

```
apijson = requests.get('https://api.census.gov/data/2015/acs/acs5/profile?get=GEO_ID,LSAD_NAME,DP03_0129PE,DP03_0129PM&in=state:27&for=county:131')
# Convert the requested json into pandas dataframe
acs5year = pd.DataFrame(columns=apijson.json()[0], data=apijson.json()[1:])
acs5year.head()
```

## Step 2: Clean Data

### 2.1: Combine 2011 and 2011-2015 data
Notice that the two dataframes have the same column names and that the desired figure has both the 1-year estimates and the 5-year estimates side-by-side. This is an indication that the two dataframes can be combined into one dataframe.

Before combining the two dataframes a new column needs to be added. The new column will identify the data as 1-year estiamte or 5-year estiamte. The new variable identifies the source, so w

```
acs1year['source'] = '1-year estimate \n (2015)'
acs5year['source'] = '5-year estimate \n (2011-2015)'
```

```
figure3_2 = pd.concat([acs1year,acs5year])
figure3_2.head()
```

The data obtained from the Census API is not in the right format. For example, right now the statistical software does not recognize the DP03_0129PE variable as a continous quantitative number.

Converting variables into the correct data type is a regular data cleaning step. The next codeblock changes the variable type to float using the astype(float) command. Floating-Point Numbers allow for decimal places and are used for continous variables. For more information on data types: https://realpython.com/python-data-types/

```
# convert estimats and MOE to float
figure3_2["DP03_0129PE"] = figure3_2["DP03_0129PE"].astype(float)
figure3_2["DP03_0129PM"] = figure3_2["DP03_0129PM"].astype(float)
```

```
import matplotlib.pyplot as plt
# Bar chart with eror
# Thanks google search "python matplotlib plot points with error bars"
## https://matplotlib.org/1.2.1/examples/pylab_examples/errorbar_demo.html

plt.figure(figsize=(10,5))
plt.errorbar(x    = figure3_2["source"],
             y    = figure3_2["DP03_0129PE"],
             yerr = figure3_2["DP03_0129PM"],
             fmt='s',
             color='b',
             markersize=8)
plt.yticks([ 0,  5, 10, 15, 20, 25],fontsize=12)
plt.xticks([-1,0,1,2],fontsize=12)
plt.ylabel('Percent of children in poverty')
plt.xlabel('Source: U.S. Census Bureau, Census API, Variables DP03_0129PE, DP03_0129PM', fontsize=8)
plt.grid(axis='y',color='k', linestyle='-', linewidth=1)
plt.suptitle('Figure 3.2. Child Poverty Rate in Rice County, Minnesota: 2015 and 2011–2015', fontsize=15)
plt.tick_params(axis='x', which='major', labelsize=12)
plt.savefig('Figure3_2.pdf', dpi=600)
```

```
years = ['2013','2014','2015','2016','2017']
acs_dict = {} # start a dictionary to store dataframes 

# loop through list of years to get data for each year
for year in years:
  for acs
  apijson = requests.get('https://api.census.gov/data/'+year+'/acs/acs1/profile?get=GEO_ID,LSAD_NAME,DP03_0129PE,DP03_0129PM&in=state:27&for=county:131')
  # Convert the requested json into pandas dataframe
  acs_dict[year] = pd.DataFrame(columns=apijson.json()[0], data=apijson.json()[1:])
  acs_dict[year]['year'] = year # add year variable to data frame

# Concatenate pandas objects into one dataframe
county_acs_df = pd.concat(acs_dict)
county_acs_df.head()
```

```
# https://data.census.gov/cedsci/table?q=C17001&g=0500000US27131&tid=ACSDT1Y2013.C17001

for year in years:
  apijson = requests.get('https://api.census.gov/data/'+year+'/acs/acs1?get=GEO_ID,LSAD_NAME,C17001_001E,C17001_001M,C17001_004E,C17001_004M,C17001_008E,C17001_008M,C17001_013E,C17001_017E&in=state:27&for=county:131')
  # Convert the requested json into pandas dataframe
  acs_dict[year] = pd.DataFrame(columns=apijson.json()[0], data=apijson.json()[1:])
  acs_dict[year]['year'] = year # add year variable to data frame

# Concatenate pandas objects into one dataframe
county2_acs_df = pd.concat(acs_dict)
county2_acs_df.head()
```

```
county2_acs_df['C17001_001E'] = county2_acs_df['C17001_001E'].astype(int)
county2_acs_df['C17001_004E'] = county2_acs_df['C17001_004E'].astype(int)
county2_acs_df['C17001_008E'] = county2_acs_df['C17001_008E'].astype(int)
county2_acs_df['C17001_013E'] = county2_acs_df['C17001_013E'].astype(int)
county2_acs_df['C17001_017E'] = county2_acs_df['C17001_017E'].astype(int)
county2_acs_df['under 18 below poverty'] = county2_acs_df['C17001_004E'] + county2_acs_df['C17001_008E']
county2_acs_df['under 18 at or above poverty'] = county2_acs_df['C17001_013E'] + county2_acs_df['C17001_017E']
county2_acs_df['under 18 total']  = county2_acs_df['under 18 below poverty'] + county2_acs_df['under 18 at or above poverty'] 
county2_acs_df['Child poverty rate'] = county2_acs_df['under 18 below poverty']/county2_acs_df['under 18 total'] 
county2_acs_df[['year','under 18 below poverty','under 18 at or above poverty','Child poverty rate']].head()
```
