# Part 3: Create a function to calculate the AAAC and AAPC projection
A function is a block of code that can be used over and over again. It is a way to make your code more efficient and easier to read. In this example, we will create a function that will calculate the projection for any given year.

```python
# what information do we need to calculate the AAAC and AAPC?
def calculate_AAAC_and_AAPC(year1, 
                            year2, 
                            pop_year1, 
                            pop_year2,
                            projection_year,
                            community_name: str = "[no name provided]"):
    """
    Calculate the annual average absolute change (AAAC) and 
    annual average percent change (AAPC) between two years.

    Parameters
    ----------
    year1 : int
        The first year of the time period.
    year2 : int
        The second year of the time period.
    pop_year1 : int
        The population in the first year of the time period.
    pop_year2 : int
        The population in the second year of the time period.
    projection_year: int
        The year to project the population to.
    community_name: str
        The name of the community. To be used in the print statement.
    """

    absolute_change = pop_year2 - pop_year1
    number_of_years = year2 - year1

    # Calculate Average Annual Absolute Change (AAAC)
    AAAC = absolute_change/number_of_years
    # Round to the nearest whole number
    AAAC_rounded = round(AAAC)
    print(f"The Average Annual Absolute Change (AAAC) is {AAAC_rounded:,}")
    # Print the result with formatted number
    print(f"For this {number_of_years} year period, " +
            f"the population of {community_name} grew by " +
            f"approximately {AAAC_rounded:,} people per year.")

    # Calculate Average Annual Percent Change (AAPC)
    AAPC = ((pop_year2/pop_year1)**(1/number_of_years)-1)
    AAPC_pct = AAPC * 100
    # Round to the nearest 2 decimal places
    AAPC_pct_rounded = round(AAPC_pct, 2)
    # Add new line
    print(f"")
    print(f"Average Annual Percent Change (AAPC) is {AAPC_pct_rounded}%")
    print(f"The population of {community_name} grew by {AAPC_pct_rounded}% " +
            f"every year between {year1} and {year2}.")

    # Calculate the projected population
    n = projection_year - year2

    # project population using AAAC
    pop_projection_AAAC = pop_year2 + n * AAAC
    # round to the nearest whole number
    pop_projection_AAAC = round(pop_projection_AAAC)

    # project population using AAPC
    pop_projection_AAPC = pop_year2 * (1+ AAPC)**n
    # round to the nearest whole number
    pop_projection_AAPC = round(pop_projection_AAPC)
    
    # Add new line
    print(f"")
    print(f"The projected population in {projection_year} using AAAC is")
    print(f"{pop_projection_AAAC:,.0f}")

    # Add new line
    print(f"")
    print(f"The projected population in {projection_year} using AAPC is")
    print(f"{pop_projection_AAPC:,.0f}")
    
    # Print summary of results
    # does the population increase or decrease?
    if (AAAC > 0) & (AAPC > 0):
        direction = "increase"
    elif (AAAC < 0) & (AAPC < 0):
        direction = "decrease"
    else:
        direction = "change"

    # Add new line
    print(f"")
    print(f"We project the population of {community_name} " +
            f"to {direction} from {pop_year2:,.0f} " +
            f"to {pop_projection_AAAC:,.0f} or " +
            f"{pop_projection_AAPC:,.0f} in {projection_year} " +
            "depending on which method we apply.")

    return AAAC, AAPC

# Check the function
AAAC, AAPC = calculate_AAAC_and_AAPC(
                        year1 = 2005, 
                        year2 = 2015, 
                        pop_year1 = 545198, 
                        pop_year2 = 649654,
                        projection_year = 2016,
                        community_name = "Denver County, Colorado")
```
